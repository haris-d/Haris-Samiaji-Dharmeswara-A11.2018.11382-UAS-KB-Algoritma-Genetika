```python
#a+2b+3c+4d = 30
#f(x) = (a+2b+3c+4d) â€“ 30

import numpy as np

# 2 Inisialisasi
n = 6

# 1 Pembentukan chromosome
chromosome = np.random.randint(0, 30,(n,4))
print("Chromosome : ", chromosome)
```

```python
epoch = 0
while epoch < 200 :
    # 3 Evaluasi Chromosome
    fungsi_objektif = abs(chromosome[:,0] + 2*chromosome[:,1] + 3*chromosome[:,2] + 4*chromosome[:,3] - 30)
    print("------------ Epoch ke ",epoch + 1)
    print("Fungsi Objektif : ", fungsi_objektif)
    print("Rata-Rata Fungsi Objektif : ", np.average(fungsi_objektif))
    
    # 4 Seleksi Chromosome
    fitness = 1/(1 + fungsi_objektif)
    print("Fitness : ", fitness)
    
    total = fitness.sum()
    print("Total Fitness : ", total)
    
    P = fitness/total
    print("Probabilitas : ", P)
    
    C = np.cumsum(P)
    print("Nilai Kumulatif Probabilitas : ", C)
    
    acak = np.random.random((chromosome.shape[0]))
    print("R : ", acak)
    
    chromosome_2 = np.zeros((chromosome.shape[0],4))
    
    for i in range(acak.shape[0]):
        for j in range(chromosome.shape[0]):
            if acak[i] < C[j]:
                chromosome_2[i,:] = chromosome[j,:]
                break
                
    chromosome = chromosome_2
    print("Chromosone diPerbarui : ", chromosome)
    
    # 5 Crossover
    R = [np.random.random() for i in range(n)]
    print("Nlai Random : ", R)
    
    pc = 0.25
    
    flag = acak < pc
    print("Nilai Flag : ", flag)
    
    cross_chromosome = chromosome[[(i == True) for i in flag]]
    len_cross_chrom = len(cross_chromosome)
    
    cross_values = np.random.randint(1,3,len_cross_chrom)
    print("C : ", cross_values)
    
    copy_chromosome = np.zeros(cross_chromosome.shape)
    for i in range(cross_chromosome.shape[0]):
        copy_chromosome[i, :] = cross_chromosome[i, :]
        
    if len_cross_chrom == 1 :
        cross_chromosome = cross_chromosome
    else :
        for i in range(len_cross_chrom):
            c_val = cross_values[i]
            if i == len_cross_chrom - 1 :
                cross_chromosome[i, c_val:] = copy_chromosome[0, c_val:]
            else :
                cross_chromosome[i, c_val:] = copy_chromosome[i+1, c_val:]
                
    print("Crossover Chromosome : ",cross_chromosome)
    
    index_chromosome = 0
    index_chromosomebaru = 0
    for i in flag :
        if i == True :
            chromosome[index_chromosome, :] = cross_chromosome[index_chromosomebaru, :]
            index_chromosomebaru = index_chromosomebaru + 1
            index_chromosome = index_chromosome + 1
            
    print("Chromosome Baru : ", chromosome)
    
    # 6 Mutasi
    a, b = chromosome.shape[0], chromosome.shape[1]
    total_gen = a*b
    print("Total generasi : ", total_gen)
    
    pm = 0.1
    jumlah_mutasi = int(np.round(pm * total_gen))
    print("Jumlah Mutasi : ", jumlah_mutasi)
    
    gen_num = np.random.randint(0, total_gen - 1, jumlah_mutasi)
    print("Generate Angka Acak : ", gen_num)
    
    ganti_angka = np.random.randint(0,30, jumlah_mutasi)
    print("Nomor Yang Akan Diganti : ", ganti_angka)
    
    for i in range(jumlah_mutasi):
        a = gen_num[i]
        baris = a//4
        kolom = a%4
        chromosome[baris, kolom] = ganti_angka[i]
    
    print("Chromosome Mutasi : ", chromosome)
    
    epoch = epoch + 1
    print("-------------------------------") 
```
